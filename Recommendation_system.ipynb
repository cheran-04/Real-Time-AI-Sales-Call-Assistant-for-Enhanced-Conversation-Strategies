# Recommendation system for Movies

1. user based
2. Item based

flow of program :
 -  movie reviews and names
 -  ratings
 -  merge both the datasets
 -  find the correlation between the movies and recommend the  movie

```python
import numpy as np
import pandas as pd
```

```python
movie_df = pd.read_csv('/content/movie.csv.zip')
```

```python
movie_df.head()
```

```python
movie_df.columns
```

```python
rating_df = pd.read_csv('/content/rating.csv')
rating_df.columns
```

```python
rating_df.head()
```

```python
df = pd.merge(movie_df,rating_df)
```

```python
df.head()
```

```python
df.shape
```

```python
pivot = df.pivot_table(index = 'userId',columns = 'title',values = 'rating')
```

```python
pivot.head(20)
```

```python
movie_watched = pivot["Toy Story (1995)"]
similarity = pivot.corrwith(movie_watched)
similarity.head()
```
